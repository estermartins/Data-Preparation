```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# Ler dados
iris = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/master/pandas/tests/data/iris.csv")
iris.head()
```

```
# Selecionando colunas 
# 1. Usando brackets de subsetting
iris[["Name", "SepalLength"]].head() 
```

```
# 2. Usando o método `loc[]`
# O método permite selecionar células dentro do banco de dados em questão utilizndo o padrão [linhas, colunas]. O método utiliza os nomes das variáveis para efetuar suas operações.
iris.loc[:,["Name", "SepalLength"]].head()
```

```
# 3. Usando o métod iloc para selecionar via index. O método `iloc` funciona de maneira  similar a loc com  diferença de que ele trabalha com os índices de linhas e variáveis. 
iris.iloc[:,[4,0]].head()
```

```
# Filtrar linhas
# 1. Usando o método .loc
# Filtrando da linha 3 até a linha 12
iris.loc[2:11, :]
```

```
# Seleciona os casos que SepalLength for > 7
iris.loc[(iris["SepalLength"] > 7),:]
```

```
# Casos SL > 6 e < 7
iris.loc[(iris["SepalLength"] > 6) & (iris["SepalLength"] < 7), :]
```

```
# 2. Query (filtra apenas linhs)
iris.query('SepalLength > 7').head()
```

```
# Combinando queries e seleções
iris.query('SepalLength > 6 % SepalLength < 7')[["SepalLength", "PetalLength", "Name"]]
```

```
iris.loc[(iris["SepalLength"] > 6) & (iris["SepalLength"] < 7), ["Name", "SepalLength", "PetalLength"]].\
sort_values(by="PetalLength")
```

```
# Criando novas vriáveis
# SepalLength > media "Size" = "Big", < media = "Small"
iris["Size"] = "Small"
m = iris["SepalLength"]. mean()
iris.loc[iris["SepalLength"] > m, "Size"] = "Big"
iris.head()
```

```
iris["Size"].value_counts()
```

```
iris["Size"].value_counts()/ iris["Size"].value_counts().sum()
```

```
fig, ax = plt.subplots()
for i in ["Small", "Big"]:
  if i == "Small":
    ax.scatter(x=iris.loc[iris["Size"] == i, "SepalLength"],
             y=iris.loc[iris["Size"] == i, "PetalLength"],
             color="green", label=i)
    else:
      ax.scatter(x=iris.loc[iris["Size"] == i, "SepalLength"],
             y=iris.loc[iris["Size"] == i, "PetalLength"],
             color="purple", label=i)
ax.legend()
plt.xlabel("SepalLength")
plt.ylabel("PetalLength")
plt.show()
```
